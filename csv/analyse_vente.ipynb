```python
import pandas as pd

vente = pd.read_csv('Ventes_Q1_2025.csv')

for i in vente :
    print(i + " :")
    print(vente[i].isna().sum())
#il n'y a des null que dans ville

venteNull = vente[(vente['ville'].isna())]
vente = vente[(-(vente['ville'].isna()))]
print(venteNull)
print(vente.head())
```

```python
for i in vente :
    liste = []
    for j in vente[i] :
        if not j in liste :
            liste.append(j)
    print(i + ' :')
    for k in liste :
        print("\t"+str(k))
```

```python
print(vente[(vente['prix_unitaire'] <= 0)])
```

```python
vente['categorie'] = vente['categorie'].apply(lambda x: "Informatique" if x == "Informatqiue" else x)

print(vente[(vente['prix_unitaire'] <= 0)]) #en observant le résultat, j'en déduit que les valeurs négatives doivent être passées au positif

vente['prix_unitaire'] = vente['prix_unitaire'].apply(lambda x : -x if x < 0 else x)
vente['quantite'] = vente['quantite'].apply(lambda x: -x if x < 0 else x)
vente['chiffre_affaires'] = vente['chiffre_affaires'].apply(lambda x: -x if x < 0 else x)
```

```python
vente["quantite"] = pd.to_numeric(vente["quantite"], errors="coerce")
vente["prix_unitaire"] = pd.to_numeric(vente["prix_unitaire"], errors="coerce")
vente["chiffre_affaires"] = pd.to_numeric(vente["chiffre_affaires"], errors="coerce")


print(vente[(vente["quantite"] == 0)])
vente["quantite"] = vente.apply(lambda row: row["chiffre_affaires"] / row["prix_unitaire"] if row["quantite"] == 0 else row["quantite"],axis=1)

print(vente[((vente["quantite"] * vente["prix_unitaire"]) != (vente["chiffre_affaires"]))])
vente["chiffre_affaires"] = vente["quantite"] * vente["prix_unitaire"]

print(vente[((vente['prix_unitaire']) != (vente['chiffre_affaires'] / vente['quantite']))])

```
